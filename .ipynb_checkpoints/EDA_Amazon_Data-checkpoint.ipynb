# EDA Project on Amazon Listing Data

## Introduction

In this exploratory data analysis (EDA) project, we will dive into a dataset that contains information about Amazon product listings. The dataset provides valuable insights into various aspects of the listings, such as product details, pricing, availability, and more. By conducting an in-depth analysis of this dataset, we aim to gain a comprehensive understanding of the Amazon product landscape and uncover interesting patterns and trends.

## Dataset Information

The dataset contains information about Amazon product listings. It includes the following columns:

- Uniq Id
- Crawl Timestamp
- Site Name
- Category
- Product Title
- Product Description
- Brand
- Pack Size Or Quantity
- Mrp
- Price
- Offers
- Combo Offers
- Stock Availability
- Image URLs

## Data Exploration

### Column Descriptions

- **Uniq Id**: A unique identifier for each product listing.
- **Crawl Timestamp**: The timestamp when the data was crawled.
- **Site Name**: The name of the website from which the data was collected.
- **Category**: The category or type of the product.
- **Product Title**: The title or name of the product.
- **Product Description**: A description of the product.
- **Brand**: The brand or manufacturer of the product.
- **Pack Size Or Quantity**: The quantity or size of the product packaging.
- **Mrp**: The maximum retail price of the product.
- **Price**: The current price of the product.
- **Offers**: Any special offers or deals available for the product.
- **Combo Offers**: Information about any combo offers associated with the product.
- **Stock Availability**: The availability status of the product.
- **Image URLs**: URLs of images associated with the product.

### What are the different categories of products available?

```python
# frequency count of product catogories
category_counts = df['Category'].value_counts()
category_counts
```

### What is the distribution of product titles across different categories?

```python
# let's create pie chart with top 7 product categories
category_counts = df['Category'].value_counts()

top_categories = category_counts.head(7)
remaining_count = category_counts[7:].sum()

plot_data = top_categories.append(pd.Series(remaining_count, index=['Other']))

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(plot_data, labels=plot_data.index, autopct='%1.1f%%')
plt.title('Top 7 Categories')
plt.show()
```

### How many products have a description available? What is the percentage?

```python
description_len = df['Product Description'].count()

percentage_with_description = (description_len / len(df)) * 100

print(f"Percentage of overall products with description: {percentage_with_description:.2f}%")
```

#### Now, let's find out percentage in each category.

```python
# Calculate the total number of descriptions in each category
category_totals = df.groupby('Category')['Product Description'].apply(lambda x: x.notnull().sum() + x.isnull().sum()).reset_index()
category_totals.columns = ['Category', 'Total Count']

# Group the data by 'Category' and count the number of non-null descriptions
category_counts = df.groupby('Category')['Product Description'].count().reset_index()
category_counts.columns = ['Category', 'Non-null Count']

# Merge the counts and totals into a single DataFrame
category_stats = pd.merge(category_counts, category_totals, on='Category')

# Calculate the percentage of non-null descriptions in each category
category_stats['Percentage'] = (category_stats['Non-null Count'] / category_stats['Total Count'].astype(float)) * 100

category_stats.sort_values('Percentage')
```

```python
# create to dataframe with in-complete categories only and since 100% is already have description
category_stats = category_stats[category_stats['Percentage'] < 100.00]
```

```python
category_stats = category_stats.sort_values("Percentage").reset_index(drop = True)
```

```python
category_stats_sorted = category_stats.sort_values(by='Percentage', ascending=False)

# Plot the data using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 10))
sns.barplot(x='Percentage', y='Category', data=category_stats_sorted)
plt.xlabel('Percentage')
plt.ylabel('Category')
plt.title('Percentage of Non-null Descriptions by Category')
plt.show()
```

### Which brands have the highest number of products in the dataset?

```python
brand_counts = df['Brand'].value_counts().nlargest(15)  # Select the top 15 brands

# Create the bar plot
plt.figure(figsize=(10, 4))
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.xticks(rotation=45)
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.title('Top Brands with Highest Number of Products')

plt.show()
```

### What is the average MRP (Maximum Retail Price) and average selling price?

```python
# Assuming all numbers will be floats (i.e. the last dot acts as a decimal point)
# you can get rid of all dots but the last dot and then cast into floats

df['Mrp'] = df['Mrp'].apply(lambda x: str(x).replace('.', '', str(x).count('.') - 1))
df['Price'] = df['Price'].apply(lambda x: str(x).replace('.', '', str(x).count('.') - 1))

df['Mrp'] = df['Mrp'].astype(float)
df['Price'] = df['Price'].astype(float)

# Calculate the average MRP and average selling price for each category
average_mrp = df.groupby('Category')['Mrp'].mean()
average_mrp.columns = ['Category', 'Mrp']

average_price = df.groupby('Category')['Price'].mean()
average_price.columns = ['Category', 'Price']
```

```python
average_stats = pd.merge(average_mrp, average_price, on='Category').reset_index()
```

```python
average_stats
```

```python
average_stats['avg_discount_percent'] = (average_stats['Mrp'] - average_stats['Price'])*100/average_stats['Mrp']
```

```python
average_stats
```

### What are the different types of offers available for the products?

```python
# Get the unique offers from the 'Offers' column
offers = df['Offers']

off_percentage = []
for offer in offers:
    if type(offer) == float:
        off_percentage.append(offer)
    elif offer[-1] == '%':
        offer = offer[:-1]
        off_percentage.append(offer)
        
df['Offers'] = pd.Series(off_percentage)
```

```python
df['Offers'] = df['Offers'].astype(float)

for column in ['Offers']:
    plt.figure()
    sns.countplot(x=pd.cut(df[column], bins=10))
    plt.xticks(rotation=45)
    plt.xlabel(column + 'in Percentage on MRP')
    plt.ylabel("Count")
```
